```python
from process_framework.references import Reference, ColumnReference
from pandas import DataFrame, Series
from typing import Any
```

```python
df_ref = Reference(DataFrame)
df_ref
```

```python
records = [
    dict(id=1, name='Simon', car='Volvo'),
    dict(id=2, name='Dave', car='Lambo')
]

df = DataFrame.from_records(records, index='id')
df
```

```python
# assign a reference to `df`
df_ref.set(df)
print('has value?:', df_ref.has_value())
df_ref.value
```

```python
# get a reference to a column in `df`
col_ref = ColumnReference(
    column='pet', 
    df=df_ref
)
col_ref
```

```python
# a series of values to assign
pets = Series({1:'dog', 2:'cat'})
pets
```

```python
col_ref.set(pets)
display('col_ref', col_ref.get_value())
display('df_ref', df_ref.value)
```

```python
fuel_ref = ColumnReference(df=df_ref, column='fuel', column_as_index='car')
fuel_ref
```

```python
fuels = Series({
    # 'Volvo':'Diesel', 
    'Lambo':'Bitcoin', 
    'Citroen':'Cidre'
})
fuels
```

```python
fuel_ref.set(fuels)
df_ref.value
```
