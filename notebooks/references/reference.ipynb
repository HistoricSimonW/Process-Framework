```python
# int
reference = Reference(int)
print(reference)

reference = Reference(int, 1)
print(reference)
```

```python
# str
reference = Reference(str)
print(reference)

reference = Reference(str, "cats")
print(reference)
```

```python
# dict
reference = Reference(dict)
print(reference)

reference = Reference(dict, {"cats":1})
print(reference)
```

```python
# type checking
for type in (int, str, dict):
    print(type, '\t', reference.is_instance_of(type))
```
