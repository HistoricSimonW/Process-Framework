```python
from process_framework.pipeline import PipelineBuilderBase, PipelineBase
```

```python
from process_framework.pipeline.metadata import RunMetadata
from process_framework.pipeline.settings import SettingsBase
from process_framework.references.reference import Reference
from process_framework.steps import Step, ModifyingStep

class Adder(ModifyingStep[int]):
    def __init__(self, subject: Reference[int], value:int, assign_to: Reference[int] | None = None):
        super().__init__(subject, assign_to)
        self.value = value
        
    def transform(self, subject: int) -> int | None:
        return subject + self.value


class Pipeline(PipelineBase):
    number:Reference[int] = Reference(int, 0)

    def get_steps(self) -> list[Step]:
        return [
            Adder(self.number, 1),
            Adder(self.number, 5)
        ]


class PipelineBuilder(PipelineBuilderBase[Pipeline]):
    def get_name(self):
        return 'what'
    
    def build_pipeline(self, settings: SettingsBase, metadata: RunMetadata) -> Pipeline:
        return Pipeline(
            name='demo pipeline',
            logging_callback=self.logging_callback
        )
    
builder = PipelineBuilder(print)
pipeline = builder.build_pipeline(None, None)
pipeline
```

```python
pipeline.execute()
```
